# Project Description


For the task you are given a dataset which can be downloaded from “https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/Concrete_Data.xls”. The dataset contains several (input) variables and one (output) variable, which are described as follows:

Cement (component 1) -- quantitative -- kg in a m3 mixture -- Input Variable

Blast Furnace Slag (component 2) -- quantitative -- kg in a m3 mixture -- Input Variable

Fly Ash (component 3) -- quantitative -- kg in a m3 mixture -- Input Variable

Water (component 4) -- quantitative -- kg in a m3 mixture -- Input Variable

Superplasticizer (component 5) -- quantitative -- kg in a m3 mixture -- Input Variable

Coarse Aggregate (component 6) -- quantitative -- kg in a m3 mixture -- Input Variable

Fine Aggregate (component 7) -- quantitative -- kg in a m3 mixture -- Input Variable

Age -- quantitative -- Day (1~365) -- Input Variable

Concrete compressive strength -- quantitative -- MPa -- Output Variable

# Data Preparation 

## Import the data

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("Concrete_Data.xls")
df.head()
```

## Describe the data

```python
df.info()
```

```python
df.describe()
```

## Visualisation of the data

```python
#plot of all the variables
for variable in df.columns:
    df[variable].plot(title=variable)
    plt.show()
    plt.clf()
```

```python
#scatter matrix to see the possible correlations
sns.pairplot(data=df,diag_kind='kde',height=5)
```

## Split the data

```python
X=df.iloc[:,0:8]
y=df.iloc[:,8]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

# Predict

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
```

```python
# Accuracy
from sklearn.metrics import  mean_absolute_error, mean_squared_error, r2_score
print("Mean absolute error: ",  mean_absolute_error(y_test, y_pred))
print("Mean squared error: ",  mean_squared_error(y_test, y_pred))
print("R2-score:", r2_score(y_test, y_pred))
```

## 10-fold cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(reg, X, y, cv=10, scoring='neg_mean_squared_error') 
print (scores)
print ("mean (negative squared error): ", scores.mean())
```

## Other regression algorithms

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import BayesianRidge, ARDRegression,LogisticRegression
from sklearn.tree import DecisionTreeRegressor
clf_names = [" Random Forest", 
             "Bayesian Ridge", 
             "ARD Regression",
             "Decision Tree"]
clfs = [ RandomForestRegressor(),
    BayesianRidge(),
    ARDRegression(),
    DecisionTreeRegressor()
    ]
```

```python
for name, clf in zip(clf_names, clfs):
    print (name)
    %time scores = cross_val_score(clf, X, y, cv=10, scoring='neg_mean_squared_error')
    print ("mean (negative squared error): ", scores.mean(), "\n")
```
